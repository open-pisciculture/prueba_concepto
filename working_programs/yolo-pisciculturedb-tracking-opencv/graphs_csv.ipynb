```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('video_data_Nov3.csv')
df["Datetime"] = pd.to_datetime(df["timestamp"])
df = df.set_index("Datetime")
del df["timestamp"]
df.head()
```

```python
df = df[df['Average Distance'] != -1]

df.plot(figsize=(20,5))
```

```python
extra_vars = df.copy()
del extra_vars['Average Distance']

extra_vars.plot(figsize=(20,5))
```
