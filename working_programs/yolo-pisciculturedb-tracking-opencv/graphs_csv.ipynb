# tinyYoloV4 Processed Video Features: Graphs

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('video_data_Nov1718_Nov2223.csv')

df["Datetime"] = pd.to_datetime(df["timestamp"])
del df['Video Index']
del df["timestamp"]

date_time_str = '19/11/21 00:00:00'
date_time_obj = datetime.strptime(date_time_str, '%d/%m/%y %H:%M:%S')
df_day1 = df[df['Datetime'] < date_time_obj]
df_day2 = df[df['Datetime'] > date_time_obj]

df = df.set_index("Datetime")
df_day1 = df_day1.set_index("Datetime")
df_day2 = df_day2.set_index("Datetime")
df.head()
```

```python
df_day1.plot(figsize=(20,20), subplots=True)
plt.savefig('video_data_img1.jpg')
```

```python
df_day2.plot(figsize=(20,20), subplots=True)
plt.savefig('video_data_img2.jpg')
```
